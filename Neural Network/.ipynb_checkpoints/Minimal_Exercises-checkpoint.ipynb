### Import the relevant libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Function to train the model and return weights, biases, and losses
def train_model_1(observations, learning_rate, loss_function):
    xs = np.random.uniform(low=-10, high=10, size=(observations, 1))
    zs = np.random.uniform(-10, 10, (observations, 1))
    inputs = np.column_stack((xs, zs))
    
    noise = np.random.uniform(-1, 1, (observations, 1))
    targets = 2 * xs - 3 * zs + 5 + noise
    
    weights = np.random.rand(2, 1)
    biases = np.random.rand(1)

    losses = []

    for i in range(100):  # Number of iterations
        outputs = np.dot(inputs, weights) + biases
        deltas = outputs - targets
        
        if loss_function == 'L2':
            loss = np.sum(deltas ** 2) / (2 * observations)  # L2 divided by 2
        elif loss_function == 'L2_no_division':
            loss = np.sum(deltas ** 2) / observations  # L2 not divided by 2
        elif loss_function == 'L1':
            loss = np.sum(np.abs(deltas)) / observations
            
        losses.append(loss)
        
        deltas_scaled = deltas / observations
        
        weights -= learning_rate * np.dot(inputs.T, deltas_scaled)
        biases -= learning_rate * np.sum(deltas_scaled)

    return xs.flatten(), zs.flatten(), targets.flatten(), outputs.flatten(), losses

# Parameters for experiments
observation_counts = [10000, 100000, 1000000]
learning_rates = [0.01, 0.001, 0.0001]
loss_functions = ['L2', 'L2_no_division', 'L1']

# Store results for each configuration
results = []

# Train the model with different configurations
for obs in observation_counts:
    for lr in learning_rates:
        for loss in loss_functions:
            xs_flattened, zs_flattened, targets_flattened, outputs_flattened, losses = train_model_1(obs, lr, loss)
            results.append({
                'Observations': obs,
                'Learning Rate': lr,
                'Loss Function': loss,
                'Inputs X': xs_flattened,
                'Inputs Z': zs_flattened,
                'Targets': targets_flattened,
                'Outputs': outputs_flattened,
                'Losses Over Iterations': losses
            })

# Convert results to DataFrame for easier analysis
results_df = pd.DataFrame(results)
results_df.head(10)
```

```python
def train_model(observations, learning_rate, loss_function, target_function):
    xs = np.random.uniform(low=-10, high=10, size=(observations, 1))
    zs = np.random.uniform(-10, 10, (observations, 1))
    inputs = np.column_stack((xs,zs))
    
    noise = np.random.uniform(-1, 1, (observations, 1))
    
    if target_function == 0:
        targets = 2 * xs - 3 * zs + 5 + noise
    elif target_function == 1:
        targets = 13 * xs + 7 * zs - 12 + noise
    
    weights = np.random.rand(2, 1)
    biases = np.random.rand(1)

    losses = []

    for i in range(100):  # Number of iterations
        outputs = np.dot(inputs, weights) + biases
        deltas = outputs - targets
        
        if loss_function == 'L2':
            loss = np.sum(deltas ** 2) / (2 * observations)
        elif loss_function == 'L2_no_division':
            loss = np.sum(deltas ** 2) / observations
        elif loss_function == 'L1':
            loss = np.sum(np.abs(deltas)) / observations
            
        losses.append(loss)
        
        deltas_scaled = deltas / observations
        weights -= learning_rate * np.dot(inputs.T, deltas_scaled)
        biases -= learning_rate * np.sum(deltas_scaled)

    return xs.flatten(), zs.flatten(), targets.flatten(), outputs.flatten(), losses, weights.flatten(), biases[0]

# Parameters for experiments
observation_counts = [10000, 100000, 1000000]
learning_rates = [0.01, 0.001, 0.0001]
loss_functions = ['L2', 'L2_no_division', 'L1']
target_functions = [0, 1]  # 0: 2*xs - 3*zs + 5, 1: 13*xs + 7*zs - 12

# User input for parameter selection
vary_param_index = int(input("Select parameter to vary (0: Observations, 1: Learning Rate, 2: Loss Function, 3: Target Function): "))
print(f"Varying parameter: {['Observations', 'Learning Rate', 'Loss Function', 'Target Function'][vary_param_index]}")

# User input for fixed parameters
if vary_param_index != 0:
    obs_index = int(input(f"Enter index for fixed observation count: [10000, 100000, 1000000] "))
    fixed_obs = observation_counts[obs_index]
    print(f"Fixed observation count: {fixed_obs}")
if vary_param_index != 1:
    lr_index = int(input(f"Enter index for fixed learning rate: [0.01, 0.001, 0.0001] "))
    fixed_lr = learning_rates[lr_index]
    print(f"Fixed learning rate: {fixed_lr}")
if vary_param_index != 2:
    loss_index = int(input(f"Enter index for fixed loss function: ['L2', 'L2_no_division', 'L1'] "))
    fixed_loss = loss_functions[loss_index]
    print(f"Fixed loss function: {fixed_loss}")
if vary_param_index != 3:
    target_index = int(input(f"Enter index for fixed target function: (0: 2*xs - 3*zs + 5, 1: 13*xs + 7*zs - 12) "))
    fixed_target = target_functions[target_index]
    print(f"Fixed target function: {['2*xs - 3*zs + 5', '13*xs + 7*zs - 12'][fixed_target]}")

# Determine which parameter to vary
if vary_param_index == 0:
    values_to_vary = observation_counts
elif vary_param_index == 1:
    values_to_vary = learning_rates
elif vary_param_index == 2:
    values_to_vary = loss_functions
else:
    values_to_vary = target_functions

print(f"\nValues to vary: {values_to_vary}")

# Prepare plots
fig, axs = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Model Performance Visualization', fontsize=16)

# Prepare data for pandas DataFrame
table_data = []

# Determine the column name based on the varying parameter
if vary_param_index == 0:
    column_name = "Observations"
elif vary_param_index == 1:
    column_name = "Learning Rate"
elif vary_param_index == 2:
    column_name = "Loss Function"
else:
    column_name = "Target Function"

# Iterate over varying parameter values
for i, value in enumerate(values_to_vary):
    if vary_param_index == 0:
        xs, zs, targets, outputs, losses, weights, bias = train_model(value, fixed_lr, fixed_loss, fixed_target)
    elif vary_param_index == 1:
        xs, zs, targets, outputs, losses, weights, bias = train_model(fixed_obs, value, fixed_loss, fixed_target)
    elif vary_param_index == 2:
        xs, zs, targets, outputs, losses, weights, bias = train_model(fixed_obs, fixed_lr, value, fixed_target)
    else:
        xs, zs, targets, outputs, losses, weights, bias = train_model(fixed_obs, fixed_lr, fixed_loss, value)
    
    # Plot Loss Over Iterations
    axs[0, i].plot(losses)
    axs[0, i].set_title(f'Loss Over Iterations\n{column_name}={value}')
    axs[0, i].set_xlabel('Iteration')
    axs[0, i].set_ylabel('Loss')
    
    # Plot Outputs vs Targets
    axs[1, i].scatter(outputs, targets, alpha=0.5)
    axs[1, i].set_title(f'Outputs vs Targets\n{column_name}={value}')
    axs[1, i].set_xlabel('Outputs')
    axs[1, i].set_ylabel('Targets')

    # Calculate accuracy (Mean Absolute Error)
    # mae = np.mean(np.abs(outputs - targets))
    def r_squared(y_true, y_pred):
        ss_res = np.sum((y_true - y_pred) ** 2)
        ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
        return 1 - (ss_res / ss_tot)

    # Calculate accuracy (R-squared)
    r2 = r_squared(targets, outputs)   

    # Add data to table
    table_data.append({
        column_name: value,
        'Weight 1': weights[0],
        'Weight 2': weights[1],
        'Bias': bias,
        #'MAE': mae
        'R-squared': r2
    })

plt.tight_layout()
plt.show()

```

```python
# Store results for each configuration
results = []

# Train the model with all possible combinations
for obs in observation_counts:
    for lr in learning_rates:
        for loss in loss_functions:
            for target in target_functions:
                xs, zs, targets, outputs, losses, weights, bias = train_model(obs, lr, loss, target)
                r2 = r_squared(targets, outputs)
                results.append({
                    'Observations': obs,
                    'Learning Rate': lr,
                    'Loss Function': loss,
                    'Target Function': target,
                    'Weight 1': weights[0],
                    'Weight 2': weights[1],
                    'Bias': bias,
                    'R-squared': r2,
                    'Final Loss': losses[-1]
                })

# Convert results to DataFrame for easier analysis
results_df = pd.DataFrame(results)

print("All configurations:")
results_df.head(10)
```

```python
# Find the best configuration based on R-squared
best_config = results_df.loc[results_df['R-squared'].idxmax()].copy()
best_config_print = best_config.copy()
best_config_print.loc["Target Function"] = ['2*xs - 3*zs + 5', '13*xs + 7*zs - 12'][int(best_config["Target Function"])]

print("\nBest configuration:")
print(best_config_print )
```

```python

# Visualize the best configuration
best_obs = best_config['Observations']
best_lr = best_config['Learning Rate']
best_loss = best_config['Loss Function']
best_target = best_config['Target Function']

xs, zs, targets, outputs, losses, weights, bias = train_model(best_obs, best_lr, best_loss, best_target)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Best Model Performance', fontsize=16)

# Plot Loss Over Iterations
ax1.plot(losses)
ax1.set_title('Loss Over Iterations')
ax1.set_xlabel('Iteration')
ax1.set_ylabel('Loss')

# Plot Outputs vs Targets
ax2.scatter(outputs, targets, alpha=0.5)
ax2.set_title('Outputs vs Targets')
ax2.set_xlabel('Outputs')
ax2.set_ylabel('Targets')

plt.tight_layout()
plt.show()
```

## Conclusion

In this series of experiments with Simple Linear Regression, we explored the effects of various parameters on model performance. Key findings include:

1. **Learning Rate Impact**: A learning rate of **0.01** yielded the best results, achieving an R-squared value of **0.996955**, indicating a strong fit to the target function.

4. **Loss Function Variations**: The L2 loss function without division performed slightly better than L1 and L2 with division, suggesting that it is more effective for this problem.
